```python
# import basic packages for data analysis

import matplotlib.pyplot as plt
import pandas as pd
import statsmodels as sm
import numpy as np
import scipy as sp
```

```python
# load neuroelectro data spreadsheet into a pandas data frame

fname = 'neurophysiology_data.xlsx'
ne_data = pd.read_excel(fname, 'Sheet1', header=0)

# rename NAcc shell and core to just MSNs
ne_data.NeuronType[ne_data.NeuronType == 'Nucleus accumbens shell neuron'] = u'Nucleus accumbens medium spiny neuron'
ne_data.NeuronType[ne_data.NeuronType =='Nucleus accumbens core neuron'] = u'Nucleus accumbens medium spiny neuron'
ne_data = ne_data.set_index('NeuronType')
```

```python
ne_data.head()
```

```python
plt.figure()
ne_data.InputResistance.plot()
```

```python
# apply simple data filtering steps (like removing data from cultured neurons)
ne_data_filt = ne_data[ne_data.Age >= 5]
ne_data_filt = ne_data_filt[ne_data_filt.PrepType != 'cell culture']
valid_species = ['Rats', 'Mice', 'Guinea Pigs']
ne_data_filt = ne_data_filt[ne_data_filt.Species.isin(valid_species)]
ne_data_filt.JxnPotential = ne_data_filt.JxnPotential.fillna(value = 'Not reported')
ne_data_filt = ne_data_filt[ne_data_filt.index != 'Neocortex uncharacterized cell']


# pool FastAhpAmplitude into AhpAmplitude
ne_data_filt.ix[ne_data_filt.AhpAmplitude.apply(np.isnan), 'AhpAmplitude'] = ne_data_filt.ix[ne_data_filt.AhpAmplitude.apply(np.isnan), 'FastAhpAmplitude']


ne_data_filt.describe()
```

```python
# get pandas data structure into a form amenable for scikit learn model fitting

# do some basic NaN filtering
ne_data_filt = ne_data_filt[ne_data_filt.SpikeHalfWidth.notnull() & ne_data_filt.Age.notnull()& ne_data_filt.ElectrodeType.notnull()& ne_data_filt.Species.notnull()]

# the y-value to be regressed against metadata
y_data = ne_data_filt.SpikeHalfWidth
#y_data = np.log10(y_data)

# setting up the x-value data matrix of features
x_data_cat_features = ['ElectrodeType','Species','Strain','JxnPotential','PrepType']
x_data = ne_data_filt[['Age', 'Temp','ElectrodeType','Species','Strain','JxnPotential','PrepType']]
x_data.loc[:,'Age'] = np.log10(x_data.loc[:,'Age'])

# figure out what the mean and medians of the x_data are
most_common_vec = x_data[['Age','Temp']].mean()
v = x_data[['ElectrodeType','Species','Strain','JxnPotential','PrepType']].mode()
n = most_common_vec.append(v.T).T
n = n.head(1)

x_data = pd.concat([x_data, n])
# x_data.tail()

s = pd.get_dummies(x_data.loc[:,'ElectrodeType'])
s2 = pd.get_dummies(x_data.loc[:,'JxnPotential'], dummy_na = True)
s3 = pd.get_dummies(x_data.loc[:,'Species'])
s4 = pd.get_dummies(x_data.loc[:,'Strain'])
s5 = pd.get_dummies(x_data.loc[:,'PrepType'])
x_data = pd.concat([x_data[['Age','Temp']], s,s2, s3,  s5], axis = 1)
x_data_col_headers = x_data.columns.values.tolist()

# split last row of most common elements out
x_data_most_common = x_data.tail(1)
x_data = x_data.iloc[:-1]
```

```python
# fit models using all data and print model coeffs

# import scikit learn model fitters
from sklearn.linear_model import LassoCV, LinearRegression, ElasticNetCV
from sklearn.ensemble import RandomForestRegressor

#kf = cross_validation.KFold(n_samples, n_folds = 5, shuffle=True, random_state = 0)
elastic_net_ob = ElasticNetCV(cv=3,fit_intercept=True, normalize=True, l1_ratio = [.1, .5, .7, .9, .95, .99, 1])
lin_reg_ob = LinearRegression(fit_intercept=True, normalize=True)
rfr = RandomForestRegressor(max_features = 'sqrt', bootstrap = True)

# run elastic net fit using all data
model = elastic_net_ob.fit(x_data, y_data)

# print elastic net model coeffs
tup = [(x_data_col_headers[i], c) for i,c in enumerate(model.coef_)]
for t in tup:
    print '%s, %0.3f ' % (t[0], t[1])

print '\n'
# run lin reg fit using all data and print coeffs
model2 = lin_reg_ob.fit(x_data, y_data)
tup = [(x_data_col_headers[i], c) for i,c in enumerate(model2.coef_)]
for t in tup:
    print '%s, %0.3f ' % (t[0], t[1])
```

```python
# run cross validated regression using different models and print accuracy
from sklearn import cross_validation

# set up cross validation object
n_samples = x_data.shape[0]
cv = cross_validation.ShuffleSplit(n_samples, n_iter=10, test_size=0.33, random_state=0)

# compute cross-validated scores
score_criteria = 'r2'
elastic_net_scores = cross_validation.cross_val_score(elastic_net_ob, x_data, y_data, cv=cv, scoring=score_criteria)
lin_scores = cross_validation.cross_val_score(lin_reg_ob, x_data, y_data, cv=cv, scoring=score_criteria)
rfr_scores = cross_validation.cross_val_score(rfr, x_data, y_data, cv=cv, scoring=score_criteria)

# this bit is to filter out inds where the linear model R^2 blows up because of overfitting
inds = np.where(np.abs(lin_scores) < 1)
if len(inds) == 0:
    inds = np.where(np.abs(elastic_net_scores) < 1)

# print mean cross-validated model accuracies on out-of-sample data

#print elastic_net_scores[inds]
print("Elastic Net Accuracy: %0.2f (+/- %0.2f)" % (elastic_net_scores[inds].mean(), elastic_net_scores[inds].std() ))

#print lin_scores[inds]
print("Ord Lin Squares Accuracy: %0.2f (+/- %0.2f)" % (lin_scores[inds].mean(), lin_scores[inds].std() ))

#print rfr_scores[inds]
print("Random Forest Accuracy: %0.2f (+/- %0.2f)" % (rfr_scores[inds].mean(), rfr_scores[inds].std() ))
```
